```python
###JOGO DA VELHA

from re import X

def tabuleiro (posicao):

    ''' Criando um tabuleiro com print e definindo a posição de onde irá cada jogada utilizando a lista de posicao'''

    print('         |          |          ')
    print('    {}    |     {}    |     {}    '.format(posicao['pos1'], posicao['pos2'], posicao['pos3']))
    print('         |          |          ')
    print('---------|----------|----------')
    print('         |          |          ')
    print('    {}    |     {}    |     {}    '.format(posicao['pos4'], posicao['pos5'], posicao['pos6']))
    print('         |          |          ')
    print('---------|----------|----------')
    print('         |          |          ')
    print('    {}    |     {}    |     {}    '.format(posicao['pos7'], posicao['pos8'], posicao['pos9']))
    print('         |          |           ')

def jogando (num_tabuleiro):

    '''definindo os numeros dos tabuleiros e perguntando em qual posição os jogadores querem posicionar os X ou O
    Além de verificar se o usuario digitou um texto ou um numero para continuar jogando (utilizando .isdigit)
    '''

    while True:
        print('\n')
        jogar = input ('Digite a posição que queira jogar: ')
        if jogar.isdigit():
            jogar = int(jogar)
            if jogar in num_tabuleiro:
                break
            
            else:
                print('Essa posição ja foi escolhida, escolha outro!')
        
        else: 
            print('Digite um numero que esteja na tabela!!!')
    return jogar

def substituindo_por_X_ou_O (jogar, num_jogadas, posicao):

    '''Substitui os numeros das posiçoes por X ou por O chamando a chave para acessar os valores'''

    for key, value in posicao.items():
        if jogar == value:
            if num_jogadas % 2 == 0:
                posicao[key] = 'X'

            else:
                posicao[key] = 'O'
        
    return posicao

def ganhar (posicao, pontuacao_X, pontuacao_Y, jogadas):

    '''fazendo todas as possibilidades de vitoria do jogador X e do jogador Y ou empate entre eles e contando o numero de vitorias do jogador que usa X e Y'''

    pontuacao_X = 0
    pontuacao_Y = 0

    K = 'X'
    J = 'O'

    if ( posicao['pos1'] == K and posicao['pos2'] == K and posicao['pos3'] == K or \
          posicao['pos4'] == K and posicao['pos5'] == K and posicao['pos6'] == K or \
          posicao['pos7'] == K and posicao['pos8'] == K and posicao['pos9'] == K or \
          posicao['pos1'] == K and posicao['pos4'] == K and posicao['pos7'] == K or \
          posicao['pos2'] == K and posicao['pos5'] == K and posicao['pos8'] == K or \
          posicao['pos3'] == K and posicao['pos6'] == K and posicao['pos9'] == K or \
          posicao['pos1'] == K and posicao['pos5'] == K and posicao['pos9'] == K or \
          posicao['pos3'] == K and posicao['pos5'] == K and posicao['pos7'] == K):

        pontuacao_X += 1
        print('O jogador do X venceu!!!')
        print('O jogador do X esta com: ', pontuacao_X)

    elif ( posicao['pos1'] == J and posicao['pos2'] == J and posicao['pos3'] == J or \
          posicao['pos4'] == J and posicao['pos5'] == J and posicao['pos6'] == J or \
          posicao['pos7'] == J and posicao['pos8'] == J and posicao['pos9'] == J or \
          posicao['pos1'] == J and posicao['pos4'] == J and posicao['pos7'] == J or \
          posicao['pos2'] == J and posicao['pos5'] == J and posicao['pos8'] == J or \
          posicao['pos3'] == J and posicao['pos6'] == J and posicao['pos9'] == J or \
          posicao['pos1'] == J and posicao['pos5'] == J and posicao['pos9'] == J or \
          posicao['pos3'] == J and posicao['pos5'] == J and posicao['pos7'] == J):

        pontuacao_Y += 1
        print('O jogador do Y venceu!!!')
        print('O jogador do Y esta com: ', pontuacao_Y)

    elif jogadas == 8:
        print('Empate')
        return 'Empate'

num_tabuleiro = [1, 2, 3, 4, 5, 6, 7, 8, 9]

posicao = {'pos1' : 1,
        'pos2' : 2,
        'pos3' : 3,
        'pos4' : 4,
        'pos5' : 5,
        'pos6' : 6,
        'pos7' : 7,
        'pos8' : 8,
        'pos9' : 9,
}
pontuacao_X = 0
pontuacao_Y = 0

for jogadas in range(9): #limitando as jogadas de 0-8 (total de 9 jogadas)
    tabuleiro(posicao) #mostra o tabuleiro
    jogar = jogando(num_tabuleiro) #utilizando a funcao de verificar se é um numero ou nao para substituir por X ou por O
    num_tabuleiro [jogar - 1] = 'X' #eliminando os numeros do tabuleiro para que nao se repita os numeros escolhidos
    posicao = substituindo_por_X_ou_O(jogar, jogadas, posicao) #substituindo os numeros por X ou por O
    if ganhar (posicao, pontuacao_X, pontuacao_Y, jogadas): #usando a funcao para ver as possibilidades e verificar quem ganhou ou empate
        tabuleiro(posicao)
        break 
    elif ganhar (posicao, pontuacao_X, pontuacao_Y, jogadas) == 'Empate':
        break
```
